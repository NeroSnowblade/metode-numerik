```python
# Wendi Kardian (2100016)
# Pendidikan Ilmu Komputer - A
# Mencari akar persamaan menggunakan Newthon Rapson Method

# Import Libraries

import matplotlib.pyplot as plt            
import numpy as np  
from sympy import *
```

```python
#create the independent variable symbol for use with sympy
x = Symbol('x')
print(x)


#fungsi f(x) yang nanti akan digunakan
user_function = x * x * x - 3*x -20

#convert into integer
userf_toprint = str(user_function)
```

```python
fig = plt.figure(num=1,figsize=(6,3),dpi=100) #ukuran untuk grafik
x_plt = np.linspace(-5,5,100)

# Grafik persamaan

# Inisialisasi canvas untuk menggambar curva grafik

# Persamaan y
y_plot = (x_plt * x_plt * x_plt )+( -3 *  x_plt)  - 20
# Turunan dari persamaan y
a_plot = (3 * x_plt * x_plt )+( -3 )
ax = fig.add_subplot(1, 1, 1)

# Inisialisasi canvas untuk grafic garis
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Berikan label pada gambar canvas yang sudah dibuat
plt.xlabel('x')
plt.ylabel('y')
plt.title('Root of:' + userf_toprint)
plt.plot(x_plt,y_plot,color='blue')
plt.plot(x_plt,a_plot,color='green')
```

```python
# Mencari turunan dari persamaan yang sudah diberikan
der_userf = user_function.diff(x)


print(der_userf)

#functions can now be used to pass numerical values

# Fungsi method lambdify untuk mengubah persamaan dalam bentuk string atau integer
# Menjadi sebuah functions persamaan yang nantinya akan digunakan

user_function = lambdify(x,user_function)
der_userf = lambdify(x,der_userf) 

```

```python
#Newton Rapson Algorithm

# Nilai x0 = 5
x = 5

#Buat list untuk nanti store nilai semua nilai x untuk di plotting datanya !
list_x_vals=[] 

# Nilai tolerance = 0.05

h = user_function(x) / der_userf(x) 


# Abs return absolute value of the number
while abs(h) >= 0.05: 
  h = user_function(x)/der_userf(x) 
  x = x - h
  list_x_vals.append(x)
  #plot setiap nilai x untuk diplotting dari setiap iterasi
  plt.plot(x,0,color='red',marker='o',markersize=5)
  print(x) 


#Plotting the final root
plt.plot(x,user_function(x),color='green',marker='o',markersize=12)

print("The value of the root is : ", "%.4f"% x)
```

```python
##################################################################
#for plotting convergence graph
fig = plt.figure(num=2,figsize=(6,3),dpi=100) #size is in inches
time = np.linspace(0,1,len(list_x_vals))
y_con = list_x_vals
plt.xlabel('t')
plt.ylabel('y')
plt.title('Convergence of: ' + userf_toprint)
# plot the function untuk proses pencarian dari setiap iterasi
plt.plot(time,y_con,color='red')
```

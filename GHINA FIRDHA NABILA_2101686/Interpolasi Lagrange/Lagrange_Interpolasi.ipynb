```python
# Nama		: Ghina Firdha Nabila
# NIM		: 2101686
# Kelas     : PILKOM 3-A
# Interpolasi

#import library numpy
import numpy as np
    
# import untuk tampilan grafik 
import matplotlib.pyplot as plt

# var n untuk memasukan banyak titik
n = int(input('Memasukan banyak titik: '))

# inisialisasi var x
x = np.zeros((n))
# inisialisasi var y
y = np.zeros((n))


# fungsi memasukan data x dan y
print('Masukan data x dan y: ')
#fungsi perulangan untuk input data
for i in range(n):
    x[i] = float(input( 'x['+str(i)+']=')) # masukan x[i]
    y[i] = float(input( 'y['+str(i)+']=')) # masukan y[i]


# var input masukan yang dicari
xp = float(input('masukan input yang mau dicari: '))

# var set data dari 0
yp = 0 

# fungsi Algoritma Lagrange
for i in range(n): 
    p = 1
    for j in range(n):
        if i != j:
            p = p * (xp - x[j])/(x[i] - x[j])
    yp = yp + p * y[i]    

# fungsi tampilan hasil
print('Pada x %.3f mempunyai f(x) atau y  %.3f' % (xp, yp))
plt.plot(x, y, marker = '*')
plt.plot(xp, yp, marker = 'o')
plt.show()
```
